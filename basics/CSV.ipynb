```python
import nltk
import csv
```

```python
sales = []
with open("sales2009.csv","rt", encoding="utf-8") as file:
    reader = csv.reader(file)
    for row in reader:
        sales.append(row)
```

```python
tokens = [nltk.word_tokenize(str(entry)) for entry in sales]
```
