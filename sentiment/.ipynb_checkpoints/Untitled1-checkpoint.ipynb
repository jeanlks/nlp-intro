```python
import nltk
import random
from nltk.corpus import names
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.style.use("ggplot")
```

```python
%matplotlib inline
```

```python
name_cfe = nltk.ConditionalFreqDist((fileid,name[-2:]) for fileid in names.fileids() for name in names.words(fileid))
```

```python
plt.figure(figsize=(50,10))
name_cfe.plot()
```

```python
def name_features(name):
    return {"pair": name[-2:]}
```

```python
name_list = [(name,"male") for name in names.words("male.txt")] + [(name,"female") for name in names.words("female.txt")]
```

```python
random.shuffle(name_list)
```

```python
features = [ (name_features(name),gender) for(name,gender) in name_list]
```

```python
len(features)/2
```

```python
training_set = features[:3972]
```

```python
testing_set = features[3972:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
classifier.classify(name_features("Brian"))
```

```python
nltk.classify.accuracy(classifier,testing_set)
```
